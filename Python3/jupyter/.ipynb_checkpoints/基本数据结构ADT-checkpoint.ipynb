## 栈 - ADT

```python
class Stack():
    def __init__(self):
        self.items = []
        
    def push(self, item):
        """入栈"""
        self.items.append(item)
        
    def pop(self):
        """出栈"""
        return self.items.pop()
    
    def peek(self):
        """弹出栈顶元素下标，不删除"""
        return len(self.items) - 1
        
    def isEmpty(self):
        """是否为空栈"""
        return self.items == []
        
    def size(self):
        """栈的数量"""
        return len(self.items)
```

```python
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(5)

print('栈顶元素下标：', stack.peek())
print(stack.isEmpty())
print(stack.size())
print(stack.pop())
stack.pop()
```

## 队列 - ADT

```python
class Queue():
    def __init__(self):
        self.items = []
        
    def enqueue(self, item):
        self.items.insert(0, item)
        
    def dequeue(self):
        return self.items.pop()
    
    def isEmpty(self):
        return self.items == []
    
    def size(self):
        return len(self.items)
```

```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue())
print(q.dequeue())
print(q.isEmpty())
q.size()
```

## 双端队列

```python
class Deque():
    def __init__(self):
        self.items = []
        
    def addFront(self, item):
        """添加到队首"""
        self.items.insert(0, item)
        
    def addRear(self, item):
        """添加到队尾"""
        self.items.append(item)
        
    def removeFront(self):
        """删除队首"""
        return self.items.pop()
        
    def removeRear(self):
        """删除队尾"""
        return self.items.pop(0)
        
    def isEmpty(self):
        return self.items == []
        
    def size(self):
        return len(self.items)
```

## 链表  

链表是一种线性表。  

链表的实现需要两种数据结构，一个是链表，一个是结点。每个结点（数据存储单元）里存放下一个结点的信息（即地址）。

```python
class Node():
    def __init__(self, item):
        self.item = item
        self.next = None
```

```python
class Link():
    def __init__(self):
        # 构造出一个空链表
        # _head存储的只能是空或者第一个结点的地址
        self._head = None
        
    def add(self, item):
        """创建一个新的结点"""
        node = Node(item)
        node.next = self._head
        self._head = node
        
    def travel(self):
        cur = self._head
        while cur:
            print(cur.item)
            cur = cur.next
            
    def isEmpty(self):
        return self._head is None
    
    def size(self):
        cur = self._head
        count = 0
        while cur:
            count += 1
            cur = cur.next
        return count
        
    def append(self, item):
        node = Node(item)
        if self._head is None:
            self._head = node
            return
        cur = self._head
        while cur.next:
            cur = cur.next
        cur.next = node
        
    def search(self, item):
        find = False
        cur = self._head
        while cur:
            if cur.item == item:
                find = True
                break
            cur = cur.next
        return find
    
    def insert(self, pos, item):
        node = Node(item)
        pre = None
        cur = self._head
        
        # 特殊情况
        if pos > self.size()-1:
            self.append(item)
            return
        if pos <= 0:
            self.add(item)
            return
        
        # 一般情况
        for i in range(pos):
            pre = cur
            cur = cur.next
        pre.next = node
        node.next = cur
        
    def remove(self, item):
        cur = self._head
        pre = None
        if cur.item is item:
            self._head = cur.next
            return
        while cur:
            pre = cur
            cur = cur.next
            if cur.item is item:
                pre.next = cur.next
                return
            
            
```

```python
link = Link()
link.add(3)
link.add(4)
link.add(5)
link.append(6)
link.append(7)
link.append(3)
# link.remove(0)
link.travel()
print(link.isEmpty())
print(link.size())
link.search(0)
```

## 树

```python
class Node():
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None
        
class Tree():
    def __init__(self):
        self.root = None
    
    def addNode(self, item):
        node = Node(item)
        if self.root is None:
            self.root = node
            return
        cur = self.root
        q = [cur] # 列表元素是我们进行遍历判断的结点
        while q:
            nd = q.pop(0)
            if nd.left is None:
                nd.left = node
                return
            else:
                q.append(nd.left)
            if nd.right is None:
                nd.right = node
                return
            else:
                q.append(nd.right)
                
    def travel(self):
        """广度遍历"""
        cur = self.root
        q = [cur]
        while q:
            nd = q.pop(0)
            print(nd.item)
            if nd.left:
                q.append(nd.left)
            if nd.right:
                q.append(nd.right)
                
    def forward(self, root):
        """前序遍历"""
        if root is None:
            return
        print(root.item)
        self.forward(root.left)
        self.forward(root.right)
        
    def middle(self, root):
        """中序遍历"""
        if root is None:
            return
        self.middle(root.left)
        print(root.item)
        self.middle(root.right)
```

```python
tree = Tree()
tree.addNode(1)
tree.addNode(2)
tree.addNode(3)
tree.addNode(4)
tree.addNode(5)
tree.addNode(6)
tree.forward(tree.root)
```

## 二叉搜索树



```python
class Node():
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None
        
class SortTree():
    def __init__(self):
        self.root = None
        
    def add(self, item):
        node = Node(item)
        cur = self.root
        if self.root is None:
            self.root = node
            return
        while cur:
            # 插入右侧
            if item > cur.item:
                if cur.right is None:
                    cur.right = node
                    break
                else:
                    cur = cur.right
            else:
                # 插入左侧
                if cur.left is None:
                    cur.left = node
                    break
                else:
                    cur = cur.left  
                    
    def forward(self, root):
        """前序遍历"""
        if root is None:
            return
        print(root.item)
        self.forward(root.left)
        self.forward(root.right)
```

```python
tree = SortTree()
alist = [3, 8, 5, 7, 6, 2, 9, 4, 1]
for i in alist:
    tree.add(i)
tree.forward(tree.root)
```

## 二分查找

* 二分查找一定是基于有序数列的查找

```python
alist = [1, 2, 3, 4, 5, 6, 7, 8, 9,]
```

```python
def find(alist, item):
    find = False
    first = 0
    mid_index = len(alist) // 2
    last = len(alist) - 1
    while first <= last:
        mid_index = first + (last - first) // 2
        if alist[mid_index] < item:
            first = mid_index + 1
        elif alist[mid_index] > item:
            last = mid_index - 1
        else:
            find = True
            break
    return find
```

```python
alist = [1, 2, 3, 4, 5, 6, 7, 8, 9,]
print(find(alist, 0))
```

## 冒泡排序

```python
def sort(alist):
    for j in range(len(alist) - 1):
        for i in range(len(alist) - 1 - j):
            if alist[i] > alist[i + 1]:
                alist[i], alist[i + 1] = alist[i + 1], alist[i]
    return alist
```

```python
alist = [3, 8, 5, 7, 6]
print(sort(alist))
```

## 选择排序

```python
def sort(alist):
    for j in range(0, len(alist) - 1):
        max_index = 0
        for i in range(1, len(alist) - j):
            if alist[max_index] < alist[i]:
                max_index = i
        alist[len(alist) - 1 - j], alist[max_index] = alist[max_index], alist[len(alist) - 1 - j]
    return alist
```

```python
alist = [3, 8, 5, 7, 6]
print(sort(alist))
```

## 插入排序

```python
def sort(alist):
    for i in range(1, len(alist)):
        while i > 0:
            if alist[i] < alist[i - 1]:
                alist[i], alist[i - 1] = alist[i - 1], alist[i]
                i -= 1
            else:
                break
    return alist
```

```python
alist = [3, 8, 5, 7, 6]
print(sort(alist))
```

## 希尔排序

* gap:
    * 增量值
    * 拆分出来的数组
* 插入排序就是增量为1的希尔排序

```python
from IPython.display import Image
Image(filename = '希尔排序.png')
```

```python
def sort(alist):
    """希尔排序基于插入排序算法"""
    gap = len(alist) // 2 # 初始gap
    while gap > 0:
        for i in range(gap, len(alist)):
            while i > 0:
                if alist[i] < alist[i - gap]:
                    alist[i], alist[i - gap] = alist[i - gap], alist[i]
                    i -= gap
                else:
                    break
        gap = gap // 2 # 缩减gap
    return alist
```

```python
alist = [3, 8, 5, 7, 6]
print(sort(alist))
```

## 快速排序

* 指定一个基数(乱序中的第一个数据值)
* 将比基数小的数据放置在基数的左侧，比基数大的数据放置在基数的右侧

```python
def sort(alist, start, end):
    if start > end:
        return
    low = start
    high = end
    mid = alist[low]
    while low < high:
        # offset hight
        while low < high:
            if alist[high] > mid:
                # 向左偏移high
                high -= 1
            else:
                alist[low] = alist[high]
                break
        # offset low
        while low < high:
            if alist[low] < mid:
                low += 1
            else:
                alist[high] = alist[low]
                break
        if low is high:
            alist[low] = mid
            break
    # 递归左侧
    sort(alist, start, low - 1)
    # 递归右侧
    sort(alist, high + 1, end)
    return alist
```

```python
alist = [6, 1, 2, 4, 9, 3, 4, 5, 10, 8]
print(sort(alist, 0, len(alist) - 1))
```
